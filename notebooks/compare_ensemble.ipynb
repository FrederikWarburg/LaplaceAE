```python
import os
import matplotlib.pyplot as plt
import numpy as np
import os
from PIL import Image
```

```python
paths = ["../figures/mnist/vae_[use_var_dec=True]/mnist_model_selection",
        "../figures/mnist/ae_[use_var_dec=True]/mnist_model_selection",
        "../figures/mnist/ae_[use_var_dec=False]/mnist_model_selection",
        "../figures/mnist/mcdropout_ae/mnist_model_selection",
        #"../figures/mnist/lae_post_hoc_[use_la_encoder=True]/mnist_model_selection",
        "../figures/mnist/lae_elbo/mnist_model_selection"] 

names = ["VAE", "AE MSE", "AE likelihood", "MC-AE", \
        #"LAE Post-hoc", \
        "LAE ELBO"]
```

```python
for name, path in zip(names, paths):
    models = os.listdir(path)
    plt.figure(figsize=(20,5))
    print(name)
    for i, model in enumerate(models):
        plt.subplot(1, len(models), i+1)
        with Image.open(f"{path}/{model}/ae_contour.png") as im:
            plt.imshow(im)
            plt.axis("off")
    plt.tight_layout()
    plt.show()
```
