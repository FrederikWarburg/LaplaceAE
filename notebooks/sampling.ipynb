```python
import matplotlib.pyplot as plt
import numpy as np
import torch
```

```python
h = np.linspace(0,100)


prior_prec = 1
sigma2 = 1 / (h + prior_prec)

plt.plot(h, sigma2,'k--', label=f"sigma2_prior_prec={prior_prec}", linewidth=5)
for prior_prec in [1, 10, 100, 1000]:
    sigma = 1 / np.sqrt(h + prior_prec) 
    plt.plot(h, sigma, label=f"sigma_prior_prec={prior_prec}", linewidth=4)
plt.xlabel("hessian")
plt.legend()
plt.show()
```

```python
h = np.linspace(0,100)


prior_prec = 1
sigma2 = 1 / (h + prior_prec)

plt.plot(h, sigma2,'k--', label=f"sigma2_prior_prec={prior_prec}", linewidth=5)
for scale in [1, 5, 10]:
    sigma = 1 / np.sqrt(h*scale + prior_prec) 
    plt.plot(h, sigma, label=f"sigma_h_scaling={scale}", linewidth=4)
plt.xlabel("hessian")
plt.legend()
plt.show()
```

```python
for init_h in range(0,20):
    h = init_h
    w = 0.1
    hs = []
    for i in range(1000):
        h = 0.99 * h + w
        hs.append(h)
    plt.plot(hs,'-', label=str(init_h))
plt.xlabel("hessian")
plt.legend()
plt.show()
```
